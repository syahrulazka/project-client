

*   Di setiap asal negara siswa (region), mayoritasnya berada di kategori umur(age_band) berapa?



```python
region_age = all_data.groupby(['region', 'age_band']).size().reset_index(name='count')
region_by_age = region_age.groupby(['region', 'age_band'])['count'].sum().reset_index()
result_pivot = region_by_age.pivot(index='region', columns='age_band', values='count')
result_pivot = result_pivot.fillna(0)
result_pivot.columns.name = None

result_pivot
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(result_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.xlabel('Age Band')
plt.ylabel('Region')
plt.title('Heatmap of Student Count by Region and Age Band')
plt.show()
```

```python
sns.catplot(
    data=all_data, y="region", hue="age_band", kind="count",
    palette="pastel", edgecolor=".6",
)
```

```python
status_siswa = all_data.groupby('final_result')['id_student'].nunique()
status_siswa = status_siswa.reset_index()
status_siswa
```

```python
cross_tab = pd.crosstab(all_data['region'], all_data['final_result'])

cross_tab.plot(kind='bar', stacked=True, colormap='Paired')

plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Relationship Between Region and Final Result')

plt.show()
```
